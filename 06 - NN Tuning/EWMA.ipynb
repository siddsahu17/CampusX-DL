```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimate.csv',  nrows=150)
```

```python
df = df[['date', 'meantemp']]
```

```python
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
```
